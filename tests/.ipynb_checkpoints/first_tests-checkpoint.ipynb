```python
def find_extreme(lst, k = 2):
    res_max = []
    res_min = []
    for i in range(k, len(lst)-k-1):
        point_max = 0
        point_min = 0
        for j in range(-k, k+1):
            if lst[i] > lst[i + j]:
                point_min += 1
            elif lst[i] < lst[i + j]:
                point_max += 1
                
        if point_min == 0:
            res_min.append(i)
        if point_max == 0:
            res_max.append(i)
    return res_max, res_min
            
```
